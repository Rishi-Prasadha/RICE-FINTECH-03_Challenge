# **Analysis Report** 
## Bitcoin Arbitration Opportunities Between Coinbase and Bitstamp, Between January 1, 2018 and March 31, 2018

---

Bitcoin markets in 2015/2016 attracted plenty of attention from arbitrage traders as prices of Bitcoin continued to increase past expectation and was getting additional media attention. This craze led traders to build their own bots at home to exploit arbitrage opportunities among Bitcoin markets during this time. As profits mounted in Bitcoin arbitration, so did corporate attention. Corporations ultimately built complex algorithmic trading bots that can exploit the smallest of arbitration opportunities. As a result, arbitration opportunities in Bitcoin were minimal to non by early 2019.

### **Assumptions** 
Given this information, since the data set that is being analyzed is between January 1, 2018 and March 31, 2018, **we are expecting to see less and less arbitration opportunities over time, with arbitration profits decreasing over time.** This is because as complex algorithmic trading bots exploit smaller and smaller opportunities, the additional buyers in cheap markets will raise Bitcoin prices here, while additional sellers in expensive markets will lower the price there. 

### **Discoveries** 
Our assumptions were correct, there were 73 arbitration opportunities that yielded greater than 1% profit in January of the data set, 3 arbitration opportunities that yielded greater than 1% profit in February, and absolutely no arbitration opportunities yielding greater than 1% in March. 

### **Calculations and Visualizations**
Using the early chosen data, January 24, 2018 as an example, the following are the calculations and visualizations done:

```python
# Create an overlay plot that visualizes the two dataframes over a period of one day early in the dataset. 
# Be sure that the plots include the parameters `legend`, `figsize`, `title`, `color` and `label` 
bitstamp.loc["2018-01-16", "Close"].plot(legend=True, figsize=(10,5), title="Bitcoin Price Bitstamp vs. Coinbase, January 16th, 2018", color="b", label="Bitstamp")
coinbase.loc["2018-01-16", "Close"].plot(legend=True, figsize=(10,5), color="pink", label="Coinbase")
```

```python
# For the date early in the dataset, measure the arbitrage spread between the two exchanges
# by subtracting the lower-priced exchange from the higher-priced one
arbitrage_spread_early = coinbase.loc["2018-01-16", "Close"]  - bitstamp.loc["2018-01-16", "Close"]  

#Calculate the spread returns by dividing the instances when the arbitrage spread is positive (> 0) 
# by the price of Bitcoin from the exchange you are buying on (the lower-priced exchange).
spread_return_early = arbitrage_spread_early[arbitrage_spread_early > 0] / bitstamp.loc["2018-01-16", "Close"]

# For the date early in the dataset, determine the number of times your trades with positive returns 
# exceed the 1% minimum threshold (.01) that you need to cover your costs
profitable_trades_early = spread_return_early[spread_return_early > 0.01]

# For the date early in the dataset, calculate the potential profit per trade in dollars 
# Multiply the profitable trades by the cost of the Bitcoin that was purchased
profit_early = profitable_trades_early * bitstamp["Close"].loc["2018-01-16"]
profit_per_trade_early = profit_early.dropna()
```

Given these calculations, the cumulative sum of the arbitration opportunities in each time slot shows that arbitration profits decrease over time. Refer to the code below (a continuation of calculations above) for January 16, February 24, and March 31, respectively:

```python
# Calculate the sum of the potential profits for the early profit per trade DataFrame
early_profit = profit_per_trade_early.sum()
middle_profit = profit_per_trade_middle.sum()
late_profit = profit_per_trade_late.sum()
display(early_profit, middle_profit, late_profit)
```
